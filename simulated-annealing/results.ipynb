```python
import numpy as np
from matplotlib import rcParams
import matplotlib.pyplot as plt
import pandas as pd
import os

rcParams['font.family'] = 'serif'
rcParams['font.size'] = 24

means_basic = []
error_basic = []
means_half = []
error_half = []
means_smart = []
error_smart = []
indexes = []

pathCWD = os.getcwd() + "/output/"


for i in range(1,11):
    indexes.append(i)
    costs_basic = []
    costs_half = []
    costs_smart = []
    for j in range(1,11):
        # ---- BASIC ----
        pathFile = pathCWD + "basic" + str(j) + "_log_" + str(i) + "s_64g_50d.dat"
        dataPkt = pd.read_csv(pathFile, sep=",")
        dfLog = pd.DataFrame(dataPkt)
        dfLog['timestamp'] = pd.to_timedelta(dfLog['timestamp'])
        dfLog['timestamp'] = [time.total_seconds() for time in dfLog['timestamp']]

        costs_basic.append([dfLog.loc[dfLog['it'] == 999999]['timestamp']])

        # ---- HALF ----
        pathFile = pathCWD + "half" + str(j) + "_log_" + str(i) + "s_64g_50d.dat"
        dataPkt = pd.read_csv(pathFile, sep=",")
        dfLog = pd.DataFrame(dataPkt)
        dfLog['timestamp'] = pd.to_timedelta(dfLog['timestamp'])
        dfLog['timestamp'] = [time.total_seconds() for time in dfLog['timestamp']]

        costs_half.append([dfLog.loc[dfLog['it'] == 999999]['timestamp']])

        # ---- SMART ----
        pathFile = pathCWD + "smart" + str(j) + "_log_" + str(i) + "s_64g_50d.dat"
        dataPkt = pd.read_csv(pathFile, sep=",")
        dfLog = pd.DataFrame(dataPkt)
        dfLog['timestamp'] = pd.to_timedelta(dfLog['timestamp'])
        dfLog['timestamp'] = [time.total_seconds() for time in dfLog['timestamp']]

        costs_smart.append([dfLog.loc[dfLog['it'] == 999999]['timestamp']])

    means_basic.append(np.mean(costs_basic))
    error_basic.append(np.std(costs_basic)/np.sqrt(len(costs_basic)))

    means_half.append(np.mean(costs_half))
    error_half.append(np.std(costs_half)/np.sqrt(len(costs_half)))

    means_smart.append(np.mean(costs_smart))
    error_smart.append(np.std(costs_smart)/np.sqrt(len(costs_smart)))


fig, ax1 = plt.subplots(figsize=(14, 6))

color1 = '#643299'
color2 = '#994032'
color3 = '#993280'
color4 = '#993242'
color5 = '#853299'
alpha = 0.6

ax1.errorbar(range(len(means_basic)), means_basic, error_basic, color=color1, mfc='white', marker='o', markersize=20,
             linestyle='', label='Basic', capsize=16, lw=4, capthick=4, zorder=6)
ax1.errorbar(range(len(means_half)), means_half, error_half, color=color3, mec='white', marker='X', markersize=20,
             linestyle='', label='Half', capsize=16, lw=4, capthick=4, zorder=6)
ax1.errorbar(range(len(means_smart)), means_smart, error_smart, color=color2, mec='white', marker='D', markersize=20,
             linestyle='', label='Smart', capsize=16, lw=4, capthick=4, zorder=6)

ax1.set_xticks(range(len(means_basic)), indexes)
ax1.set_xlabel("Problem Instance")
ax1.set_ylabel("Solving Time")
ax1.tick_params(axis='y', which='major')
ax1.set_ylim(60)
ax1.grid(True, 'major', 'y', linestyle='--', color='darkgrey', zorder=0)

bars, labels = ax1.get_legend_handles_labels()
ax1.legend(bars, labels, loc='best')
# fig.suptitle('Solving Time by Number of Virtual Positions')
# plt.show()
# plt.savefig('fig/fig..pdf', format='pdf', bbox_inches = 'tight', pad_inches = 0)
```
